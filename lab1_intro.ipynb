## 1. Plot data using `matplotlib`


See [here](https://matplotlib.org/stable/tutorials/index.html) for official `matplotlib` tutorial

```python
# standard package for numerical calculations  
import numpy as np                 

# standard package for plotting!
import matplotlib.pyplot as plt                   
```

```python
# generate 100 samples from normal distribution
x = np.random.randn(100)
y = np.random.randn(100)
```

```python
# simple
plt.plot(x,y, '.')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of X vs. Y')
```

```python
# the scatter plot
plt.scatter(x,y, s=20, alpha=0.8)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter plot of X vs. Y')
```

## 2. Load data using `pandas`

Good references on `pandas`: 
* [Officeial pandas tutorial](https://pandas.pydata.org/pandas-docs/stable/getting_started/tutorials.html)

* [10 Minutes to pandas](https://pandas.pydata.org/pandas-docs/stable/10min.html)



```python
# let's import pandas
import pandas as pd
```

```python
# read Auto data
Auto = pd.read_csv('data/Auto.csv', na_values='?')  # => DataFrame
# show only head
print(Auto.shape)                    # size = 397 x 9
Auto.head()
```

```python
# an example of missing data
Auto.loc[32]
```

```python
# take care of missing data
Auto = Auto.dropna()
print(Auto.shape)     # now we have 392 x 9
```

```python
# get column names
list(Auto)
# equivalently, 
#Auto.columns.values.tolist()
```

```python
# get a column => Series
Auto['mpg']
```

```python
# plot histogram of mpg
plt.hist(Auto['mpg'])
```

```python
# plot cylinders vs mpg
plt.plot(Auto['cylinders'], Auto['mpg'], '.', alpha=0.2)
```

```python
#Equivalently
Auto.plot.scatter(x='cylinders', y='mpg', alpha=0.2)
```

```python
# use "group-by" to plot boxplot for each cylinder types
Auto.boxplot(column='mpg', by='cylinders')
```
